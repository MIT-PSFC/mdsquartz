````python
from MDSplus import Tree
import matplotlib.pyplot as plt
````

````python
tree=Tree('magtest', 1)
````

````python
n = tree.QUARTZ.PCS_MAGNETIC_CONTROL.COIL_CURRENT_TARGET_CS1L
sig = n.evaluate()
````

````python
plt.plot(sig.dim_of().data()[0:10000],sig.data()[0:10000])
plt.show()
````
